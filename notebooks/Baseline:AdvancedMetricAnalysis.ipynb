# **Baseline/Advanced Modeling Metric Analysis**

Analyzing and comparing the performance metrics of the two modeling approaches used—Baseline (Ridge Regression) and Advanced (Gradient Boosting)—to evaluate their effectiveness in predicting fantasy football player performance. The goal is to assess how well each model captures trends and variability in the data while maintaining generalizability to unseen scenarios. The Baseline Model (Ridge Regression) serves as a simple, regularized linear regression approach designed to capture general trends without overfitting. The Advanced Model (Gradient Boosting) leverages decision tree ensembles to capture complex, non-linear patterns in the data. This analysis will provide critical insights into the strengths and limitations of each modeling approach as well as the challenges of making accurate predictions in this context.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing libraries/packages

```python
baselineEvalData = {
    'Position': ['Quarterback', 'Running Back', 'Wide Receiver', 'Tight End'],
    'R-squared': [0.912, 0.797, 0.828, 0.753],
    'MAE': [2.04, 2.36, 2.28, 1.74]
}
#creating dictionary with baseline evaluation data

evaldf = pd.DataFrame(baselineEvalData)
#turning the baseline evaluation data into a dataframe

df_melted = evaldf.melt(id_vars='Position', var_name='Metric', value_name='Value')
#melting the dataframe to use seaborn

plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Position', y='Value', hue='Metric', palette='colorblind')
plt.title('Ridge Regression Model Performance by Position', fontsize=16)
plt.xlabel('Positions', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Metric', fontsize=10)
plt.tight_layout()
plt.show()
#plotting
```

```python
advancedEvalData = {
    'Position': ['Quarterback', 'Running Back', 'Wide Receiver', 'Tight End'],
    'R-squared': [0.9242, 0.9253, 0.970084, 0.9465],
    'MAE': [1.819, 0.959, 0.6255, 0.599]
}
#creating dictionary with advanced evaluation data

evaldf = pd.DataFrame(advancedEvalData)
#turning the advanced evaluation data into a dataframe

df_melted = evaldf.melt(id_vars='Position', var_name='Metric', value_name='Value')
#melting the dataframe to use seaborn

plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Position', y='Value', hue='Metric', palette='colorblind')
plt.title('Gradient Boosting Model Performance by Position', fontsize=16)
plt.xlabel('Positions', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Metric', fontsize=10)
plt.tight_layout()
plt.show()
#plotting
```

When comparing the baseline Ridge Regression model to the advanced Gradient Boosting model, the differences in performance metrics across positions highlight the trade-offs between simplicity and complexity.

**Ridge Regression Performance**
- 
**R-Squared**
- Ridge Regression achieves moderate R-squared values across all positions, ranging from 0.753 (Tight End) to 0.912 (Quarterback).
- This indicates that Ridge Regression captures general trends and explains a decent amount of variance in fantasy points, but it does not overly rely on specific patterns.

**Mean Absolute Error (MAE)**
- The MAE values for Ridge Regression are higher and more realistic, with 2.04 for Quarterbacks and 1.74 for Tight Ends.
- This reflects the model's simplicity, as it does not capture intricate relationships in the data, which may lead to less precise predictions.

**Conclusion**
- The Ridge Regression model is less likely to overfit, making it more robust and reliable for generalization, especially given the high variability of football player performance.

**Gradient Boosting Performance**
- 
**R-squared**
- Gradient Boosting achieves exceptionally high R-squared values, averaging at around ~94%
- While this suggests the model explains a hihg amount of variance in fantasy points, it raises concerns about overfitting, as the R-squared values are unrealistically high for inherently unpredictable data.

**Mean Absolute Error (MAE)**
- The MAE values are much lower compared to Ridge Regression, with 1.819 for Quarterbacks and 0.599 for Tight Ends.
- These unrealistically low MAE values for positions like Wide Receivers (0.6255) and Tight Ends (0.599) also suggest the Gradient Boosting model is overfitting by memorizing the training data rather than learning generalizable patterns.

**Conclusion**
- While Gradient Boosting appears to outperform Ridge Regression based on raw metrics, the model's extremely low MAE and high R-squared values likely stem from overfitting. This is particularly evident for Wide Receivers and Tight Ends, whose week-to-week performance is highly variable and cannot be predicted with such precision.


**Comparison and Takeaways**
- 
**Overfitting in Gradient Boosting**
- The Gradient Boosting model's performance metrics indicate it is overly complex for this dataset, capturing noise and spurious patterns instead of meaningful trends.
- For fantasy football data, where variability and external factors (e.g., game context, injuries) are significant, such high precision is unrealistic.

**Why Ridge Regression is Better**
- Ridge Regression provides more realistic and generalizable results, with higher MAE values that better reflect the unpredictable nature of player performance.
- Its simplicity ensures the model does not overfit and can perform reliably on unseen data.

By utilizing Ridge Regression, the predictions align better with the inherent variability of the data, making it a more practical and trustworthy choice.
